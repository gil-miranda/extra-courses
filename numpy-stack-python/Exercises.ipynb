### Ex 1
Itere v 25 vezes, tal que
$$
v' = vA\\
v = v'
$$

Plote a dist√¢ncia euclidiana entre $v'$ e $v$

```python
A = np.array([[0.3,0.6,0.1],[0.5,0.2,0.3],[0.4,0.1,0.5]])
v = np.array([1/2,1/4,1/4])
```

```python
dist = []
for i in range(0,25):
    v_next = v*A
    e_d = sum(sum(v_next**2 + v**2))
    dist.append(e_d)
    v = v_next
```

```python
plt.plot(dist)
plt.show()
```

```python
c = list(map(len,x)) ## count of ocurrences for each number
```

```python
m = [[0]]*10
for i in range(0,10):
    m[i] = sum(x[i][:].values)/c[i]
```

```python
#plt.imshow()
fig, ax = plt.subplots(5,2, figsize=(5,10))
for i in range(0,5):
    for j in range(0,2):
        ax[i][j].imshow(255 - m[2*i+j],cmap='gray')
```

### Ex 4
Escrever uma fun√ß√£o para determinar se uma matriz √© sim√©trica

Def.: Seja A uma matriz, A √© sim√©trica se:
$$
A = A^T
$$

```python
def is_sqr(A):
    if A.shape[0] == A.shape[1]:
        return True
    else:
        return False
def is_symm(A):
    if is_sqr(A):
        if np.array_equal(A,A.T):
            return True
        else:
            return False
    else:
        return False
```

```python
A = np.array([[9,13,5,2],[12,3,60,23],[22,22,3,7],[0,0,0,0]])
A
```

```python
is_symm(A)
```

```python
is_symm(B)
```
