```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

#### Lineplot

```python
x = np.linspace(0,10,1000)
y = np.sin(x)
plt.plot(x,y, label='Velocity')
plt.xlabel('Time')
plt.ylabel('Distance')
plt.title('My cool chart')
plt.legend()
plt.show()
```

### Scatterplot

```python
A = pd.read_csv('datasets/data_1d.csv', header=None).values
```

```python
x = A[:,0]
y = A[:,1]
```

```python
plt.scatter(x,y)
```

```python
x_line = np.linspace(0,100,100)
y_line = 2*x_line+1

plt.scatter(x,y)
plt.plot(x_line, y_line)
```

### Histogram

```python
plt.hist(x)
```

```python
R = np.random.random(10000)
plt.hist(R)
```

```python
plt.hist(R,bins=20)
```

```python
y_actual = 2*x + 1
residuals = y - y_actual
plt.hist(residuals)
```
